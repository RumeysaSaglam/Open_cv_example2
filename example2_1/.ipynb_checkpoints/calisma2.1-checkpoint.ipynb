```python
import cv2
import numpy as np
%matplotlib inline
img = cv2.imread('racecar-j.jpg')
img.shape
import matplotlib.pylab as plt
x_start = 0
x_end = 10
y_start = 0
y_end = 10
channel = 0
img[y_start:y_end,x_start:x_end,0]
plt.axis("off")
plt.imshow(img)
plt.show()

left_up = img[0:145,0:250,:]
right_up = img[0:145,250:500,:]
left_down = img[0:145,500:750,:]
right_down = img[0:145,750:1000,:]
left_up1 = img[145:290,0:250,:]
right_up1 = img[145:290,250:500,:]
left_down1 = img[145:290,500:750,:]
right_down1 = img[145:290,750:1000,:]
left_up2 = img[290:435,0:250,:]
right_up2 = img[290:435,250:500,:]
left_down2 = img[290:435,500:750,:]
right_down2 = img[290:435,750:1000,:]
left_up3 = img[435:580,0:250,:]
right_up3 = img[435:580,250:500,:]
left_down3 = img[435:580,500:750,:]
right_down3 = img[435:580,750:1000,:]

image_parts = [left_up,right_up,left_down,right_down,left_up1,
right_up1,left_down1,right_down1,left_up2,right_up2,left_down2,
               right_down2,left_up3,right_up3,left_down3,right_down3]

for part_img in image_parts:   
    plt.axis("off")
    plt.imshow(part_img)
    plt.show()
```

```python
swapped_image = np.zeros_like(img)
```

```python
swapped_image[0:145,0:250,:]=right_up
swapped_image[0:145,250:500,:]=right_down
swapped_image[0:145,500:750,:]=left_up
swapped_image[0:145,750:1000,:]=right_up2
swapped_image[145:290,0:250,:]=right_down1
swapped_image[145:290,250:500,:]=left_up1
swapped_image[145:290,500:750,:]=right_up3
swapped_image[145:290,750:1000,:]=left_down
swapped_image[290:435,0:250,:]=left_down1
swapped_image[290:435,250:500,:]=right_down3
swapped_image[290:435,500:750,:]=left_up2
swapped_image[290:435,750:1000,:]=left_down2
swapped_image[435:580,0:250,:]=left_down3
swapped_image[435:580,250:500,:]=left_up3
swapped_image[435:580,500:750,:]=right_up1
swapped_image[435:580,750:1000,:]=right_down2
```

```python
plt.axis('off')
plt.imshow(swapped_image)
plt.show()
cv2.imwrite('swapped.jpg',swapped_image)
```
